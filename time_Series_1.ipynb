Step-by-Step Example: Multiplicative Decomposition

Import Necessary Libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Set the seaborn style
sns.set(style='whitegrid')
```

Create a Sample Time Series Data

```
# Generate a time index for 3 years of monthly data
np.random.seed(42)
date_rng = pd.date_range(start='2019-01-01', end='2021-12-01', freq='MS')

# Generate synthetic sales data with trend and seasonality
trend = np.linspace(50, 200, len(date_rng))
seasonality = 10 + 5 * np.sin(2 * np.pi * date_rng.month / 12)
noise = np.random.normal(0, 5, len(date_rng))
sales_data = trend * seasonality + noise

# Create a DataFrame
data = pd.DataFrame({'Date': date_rng, 'Sales': sales_data})
data.set_index('Date', inplace=True)
```

Plot the Time Series Data

```
# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(data, label='Sales')
plt.title('Monthly Sales Data')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

Interpret the Decomposed Components

```
# Perform multiplicative decomposition
result = seasonal_decompose(data['Sales'], model='multiplicative', period=12)

# Plot the decomposed components
result.plot()
plt.show()
```

# New Section

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Setting the seaborn style
sns.set(style='whitegrid')

# Generate a time index for 3 years of monthly data
np.random.seed(42)
date_rng = pd.date_range(start='2019-01-01', end='2021-12-01', freq='MS')

# Generate synthetic sales data with trend and seasonality
trend = np.linspace(50, 200, len(date_rng))
seasonality = 10 + 5 * np.sin(2 * np.pi * date_rng.month / 12)
noise = np.random.normal(0, 5, len(date_rng))
sales_data = trend + seasonality + noise

# Create a DataFrame
data = pd.DataFrame({'Date': date_rng, 'Sales': sales_data})
data.set_index('Date', inplace=True)

# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(data, label='Sales')
plt.title('Monthly Sales Data')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Perform additive decomposition
result = seasonal_decompose(data['Sales'], model='additive', period=12)

# Plot the decomposed components
result.plot()
plt.show()
```
